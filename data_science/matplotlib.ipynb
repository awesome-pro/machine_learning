```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('sales_data.csv')

# Display the first few rows of the dataset
df.info()
```

```python
total_sales_by_product = df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product.head()
```

```python
total_sales_by_product.plot(kind='bar', color='teal').set_title('Total Sales by Product Category')
```

```python
# plot sales trend over time
sales_trend = df.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'], sales_trend['Total Revenue'])
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.title('Sales Trend Over Time')
plt.show()
```
