```python
import pandas as pd

data = [1,2, 3, 4, 5]

series = pd.Series(data)

print(series)
print(type(series))
```

```python
data= { "a": 1, "b": 2, "c": 3}
series = pd.Series(data)
print(series)
print(type(series))
print(series.index)
print(series.values)
```

```python
# DataFrames
# create a DataFrame from dictionary

data = {
    "name": ["John", "Anna", "Peter", "Linda"],
    "age": [28, 22, 34, 42],
    "city": ["New York", "Paris", "Berlin", "London"]
}
df = pd.DataFrame(data)

# df.iloc[0]
# df.at[2, 'city']
df.iat[0, 2]
```

```python
# Create from list of dictionaries

data = [
    {"name": "John", "age": 28, "city": "New York"},
    {"name": "Anna", "age": 22, "city": "Paris"},
    {"name": "Peter", "age": 34, "city": None},
    {"name": "Linda", "age": 42, "city": "London"}
]
df = pd.DataFrame(data)
print(df)
```
